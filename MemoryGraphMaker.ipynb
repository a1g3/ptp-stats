```python
import matplotlib.pyplot as plt

# Data points (Elapsed time in minutes/seconds and Memory consumed in KB)
elapsed_time = ['45 sec', '21 min', '51 min', '6hr']
memory_consumed = [123.6, 125.4, 125.5, 125.5]

# Convert elapsed time to seconds for easier plotting
time_in_seconds = [45, 21 * 60, 51 * 60, 6 * 3600]

# Set the figure size (width, height) in inches
plt.figure(figsize=(20, 10))

# Plot the data, connecting the dots with a line
plt.plot(time_in_seconds, memory_consumed, marker='o', color='r', label='Memory Usage')

# Fill the area beneath the plot with a color (red in this case)
plt.fill_between(time_in_seconds, memory_consumed, color='red', alpha=0.3)

# Customize the x-axis to show the time in a readable format (minutes and seconds)
plt.xticks(time_in_seconds, elapsed_time)

# Add labels and title
plt.xlabel('Elapsed Time')
plt.ylabel('Memory Consumed (KB)')
plt.title('Memory Consumption Over Time')

# Show the plot
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data points (Elapsed time in minutes/seconds and Memory consumed in KB)
elapsed_time = ['0', '0.5', '21 min', '51 min', '6hr']
memory_consumed = [0, 123.6, 125.4, 125.5, 125.5]

# Convert elapsed time to seconds for easier plotting
time_in_seconds = [0, 45, 21 * 60, 51 * 60, 60 * 60]  # Corrected 6 hr to 6 * 3600

# Set the figure size (width, height) in inches
plt.figure(figsize=(20, 10))

# Plot the first part (up to 51 min)
time_part_1 = time_in_seconds[:4]
memory_part_1 = memory_consumed[:4]

# Plot the second part (from 6hr onward)
time_part_2 = time_in_seconds[4:]
memory_part_2 = memory_consumed[4:]

# Plot the first part of the data (up to 51 min)
plt.plot(time_part_1, memory_part_1, marker='o', color='r', label='Memory Usage (Part 1)')

# Plot the second part of the data (from 6hr onward)
plt.plot(time_part_2, memory_part_2, marker='o', color='r', linestyle='-', label='Memory Usage (Part 2)')

# Fill the area beneath the plot with a color (red in this case)
plt.fill_between(time_part_1, memory_part_1, color='red', alpha=0.3)
plt.fill_between(time_part_2, memory_part_2, color='red', alpha=0.3)

# Manually create the break (gap between 51 min and 6 hours)
# Plot a dashed line to represent the gap
plt.plot([max(time_part_1), min(time_part_2)], [max(memory_part_1), max(memory_part_1)], color='black', lw=3, linestyle='--')

# Add break markers manually (slashes) to show the axis break
plt.text(60*60, 100, r'$\backslash$', fontsize=20, ha='center')
plt.text(1.2 * max(time_part_1), 100, r'$\backslash$', fontsize=20, ha='center')

# Add additional slashes for the top/bottom of the break for prominence
plt.text(1.2 * max(time_part_1), 100, r'$\backslash$', fontsize=20, ha='center')
plt.text(1.2 * max(time_part_1), 100, r'$\backslash$', fontsize=20, ha='center')

# Customize the x-axis to show the time in a readable format (minutes and seconds)
plt.xticks(time_part_1 + time_part_2, elapsed_time)

# Add labels and title
plt.xlabel('Elapsed Time')
plt.ylabel('Memory Consumed (KB)')
plt.title('Memory Consumption Over Time (Broken Axis)')

# Show the plot
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data points (Elapsed time in minutes/seconds and Memory consumed in KB)
elapsed_time = ['', '0.5', '21.5', '52', '360']
memory_consumed = [0, 123.6, 125.4, 125.5, 125.5]

# Convert elapsed time to seconds for easier plotting
time_in_seconds = [0, 45, (21 * 60)+38, (51 * 60)+51, 60*60]

# Set the figure size (width, height) in inches
plt.figure(figsize=(20, 10))

# Plot the data, connecting the dots with a line
plt.plot(time_in_seconds, memory_consumed, marker='o', color='r', label='Memory Usage')

# Fill the area beneath the plot with a color (red in this case)
plt.fill_between(time_in_seconds, memory_consumed, color='red', alpha=0.3)

fontsize_value = 28

# Customize the x-axis to show the time in a readable format (minutes and seconds)
plt.xticks(time_in_seconds, elapsed_time, fontsize=fontsize_value)  # Increased fontsize for x-ticks
plt.yticks(fontsize=fontsize_value)
# Add break markers manually (slashes) to show the axis break
plt.text(55.5*60, 123.5, r'$/$', fontsize=fontsize_value, ha='center')
plt.text(55*60, 123.5, r'$/$', fontsize=fontsize_value, ha='center')

# Add break markers manually (slashes) to show the axis break
plt.text(55.5*60, -3, r'$/$', fontsize=fontsize_value, ha='center')
plt.text(55*60, -3, r'$/$', fontsize=fontsize_value, ha='center')

# Set the limits of the axes to avoid negative values
plt.xlim(0, 3601)  # A little extra space on the right
plt.ylim(0, max(memory_consumed) * 1.1)  # A little extra space on the top

for i, v in enumerate(memory_consumed):
    if (i == 0):
        continue
    plt.text(time_in_seconds[i], v+5, "%s" %v, ha="center", fontsize=24)

# Add labels and title
plt.xlabel('Elapsed Time (min)', fontsize=fontsize_value)
plt.ylabel('Memory Consumed (KB)', fontsize=fontsize_value)

# Show the plot
plt.grid(True)
plt.show()
```
